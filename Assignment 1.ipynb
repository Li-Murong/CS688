```python
from sklearn import datasets
import pandas as pd
```

#### 1. Read wine dataset

```python
wine = datasets.load_wine()
```

#### 2. Print and review the 'DESCR' of this dataset.

```python
print(wine.DESCR)
```

#### 3. Plot histograms of all features

```python
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_df.head()
```

```python
from matplotlib import pyplot as plt

wine_df.hist(bins=15, figsize=(15, 10), layout=(4, 4))
plt.tight_layout()
plt.show()
```

#### 4. Use dataframe.describe to get table

```python
wine_df.describe()
```

```python
wine_df.quantile([.1, .9])
```

#### 5. Are there any null values in these features?
#### Answer: No null values.

```python
null_values = wine_df.isnull().sum()
null_values_df = pd.DataFrame(null_values, columns=['Null Counts'])

null_values_df
```

#### 6. Split the dataset into .2 VS .8, then use random forest classifier to get accuracy on test sets.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = wine_df
y = wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=688)

clf = RandomForestClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy
```
